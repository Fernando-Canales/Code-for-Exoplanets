# Playing with parameters of a RV plot (II)

## In this time I will play with the parameters of the radial velocity function from <a href="https://arxiv.org/pdf/1009.1738.pdf">Murray and Correira's</a>  article (or book chapter). The equation is the following:

$$v_{r} = v_{z} + K ( cos(\omega + f) + ecos \omega )$$

## Where:

$$K = \frac{m_{2}}{m_{1}+m_{2}} \frac{nasin(I)}{\sqrt{1-e^{2}}}$$

## As can be seen, there isn't an explicit dependence on time $t$ in the equations. So let's make some calculations:

$$v_{r} = v_{z} + K[ cos(\omega)cos(f) -sin(\omega)sin(f) + ecos(\omega)] $$

## Now we remember that the mean anomaly, $M$, can be converted into the true anomaly, $f$, (angle between periapsis and actual position of the body). Moreover, is convenient to put $\omega = \frac{\pi}{2}$. as Oscar did (I know this is obviously convenient but I don't understand why we can put this value to $\omega$). However, I will mantain $\omega$ as a free parameter:

$$v_{r} = v_{z} + K[ cos(\omega)cos(M) -sin(\omega)sin(M) + ecos(\omega)] $$

## Now let's remember the value of the mean anomaly:

$$M = n(t-t_{0}) $$

## Where $n = \frac{2 \pi}{T}$ is the mean motion, where $T$ is the orbital period and $t_{0}$ is the time of periastron passage. Then:

$$ v_{r} = v_{z} + K[ cos(\omega)cos(\frac{2\pi(t-t_{0})}{T}) -sin(\omega)sin(\frac{2\pi(t-t_{0})}{T}) + ecos(\omega)]$$ 

## As can be seen, if we put $\omega = \frac{\pi}{2}$ in the expression above, we obatin Oscar's expression:

$$v_{r} = v_{z} - Ksin(\frac{2\pi(t-t_{0})}{T})$$

```python
#Import main libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#We write the funcition:
#K  -> Doppler semi amplitude
#vZ -> Proper motion of the barycenter (or star's systemic velocity)
#w  -> argument of periapsis
#e  -> Excentricity 
#T  -> Orbital period
#t0 -> Time of passage of periastron (or time of minimal conjunction)
def calcula_RV(t,parametros):
    vz = parametros[0]
    K  = parametros[1]
    T  = parametros[2]
    w  = parametros[3]
    t0 = parametros[4]
    e  = parametros[5]
    vr = vz + K*(np.cos(w)*np.cos(2*np.pi*(t-t0)/T) - np.sin(w)*np.sin(2*np.pi*(t-t0)/T) + e*np.cos(w))
    return vr

#Now the time vector will be displayed. I will put it from 0 to 20 days, with a point in the plot every 0.2 days.

t = np.arange(0,20,0.2)

#Now the parameters

vz = -40 #km/s
K  = 0.7 #km/s
T  = 2   #days
w  = 33  #Â°
t0 = 13  #days
e  = 0.001

#Now we call the function that plots the RV plot

vr = calcula_RV(t,[vz,K,T,w,t0,e])

#Now we plot it

plt.plot(t,vr)
```
