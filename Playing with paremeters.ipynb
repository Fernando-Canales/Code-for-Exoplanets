# Playing with the parameters of a RV plot.

## I will change sistematically the parameters of the RV plot that was made in the couse given by Oscar. Those parameters are (vz, K,P,t0). They are described as follows. However, the expresion that compute the radial velocity function is still not very clear for me tbh: 
$$  v_{r} = v_{z} - K sin\left( \frac{2 \pi (t-t_{0})}{P} \right)$$
## How is it derived?

```python
#Import the main libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Define the function
#The paramerters art the following:
#vz -> the proper motion of the barycenter or Oscar pointed out, "star's systemic velocity". 
#K  -> Doppler semi-amplitude
#P  -> Orbital period of the star around the barycenter
#t0 -> Time of minimum conjunction (I don't undestand this parameter, is the time of periastron passage?)
def calcula_RV(t,params):
    vz = params[0]
    K  = params[1]
    P  = params[2]
    t0 = params[3]
    vr = vz - K * np.sin(2*np.pi*(t-t0)/P) # I don understand either how this equation is obtained.
    return vr
#Now the time vector. I will put it from 0 to 20 days, with a point in the plot every 0.2 days.
t = np.arange(0,20,0.2)
#Now we give values to the parameters
vz = 100 #km/s
K  = 0.5   #km/s
P  = 2   #days
t0 = 50 #days
#We call the function
vr = calcula_RV(t, [vz,K,P,t0])
plt.plot(t,vr)
```

## As can be seen, there are 10 peaks in the plot. Since the observation time in this case was 20 days and the orbital period is 2 days, 10 orbital periods is the otuput. Therefore, as we increase the time of observations, the plot will show more peaks. On the other hand, if we increase the duration of the orbital period, then less peaks will appear in the plot. Also, I put $v_{z} = 100 \frac{km}{s}$ and $K = 0.5 \frac{km}{s}$. This means that the "y-axis"  is centered at 100 and its maximum value is 100.5 ($v_{z} + K$) and its minimum value is 99.5 ($v_{z} - K$).  However, the parameter $t_{0}$ seems to have any influence at all on the plot. Let's change it.

```python
#Now the time vector. I will put it from 0 to 20 days, with a point in the plot every 0.1 days.
t = np.arange(0,20,0.1)
#Now we give values to the parameters and change t0 to 500 days
vz = 100 #km/s
K  = 0.5   #km/s
P  = 2   #days
t0 = 500 #days
#We call the function
vr = calcula_RV(t, [vz,K,P,t0])
plt.plot(t,vr)
```

## I went from $t_{0} = 50$ days to $t_{0} = 500 $ days in the last plot and it seems that anything happened. Why? Let's make $t_{0}$ negative or something like that.

```python
#Now the time vector. I will put it from 0 to 20 days, with a point in the plot every 0.1 days.
t  = np.arange(0,20,0.1)
#Now we give values to the parameters
vz = 100 #km/s
K  = 0.5   #km/s
P  = 2   #days
t0 = -5 #days. This of course has no physical meaning.
#We call the function
vr = calcula_RV(t, [vz,K,P,t0])
plt.plot(t,vr)
```

## Now there is a change! The plot this time starts growing instead of de-growing like in the first two plots.  That certainly has something to do with the "-" sign of $t_{0}$. But since the value $t_{0}$ is a physical time ( although certainly I don't understand its meaning very well), it has to be always non-negative. So let's make it positive again and see how this can affect the  plot with a small value.

```python
#Now the time vector. I will put it from 0 to 20 days, with a point in the plot every 0.1 days.
t = np.arange(0,20,0.1)
#Now we give values to the parameters
vz = 100 #km/s
K  = 0.5   #km/s
P  = 2   #days
t0 = 0.5 #days 
#We call the function
vr = calcula_RV(t, [vz,K,P,t0])
plt.plot(t,vr)
```

## A value of $t_{0} = 0.5$ days changes both the initial and final points of the plot! As can be seen, $t_{0}$ is acting like a trigonometric phase for the sine function, i.e. $sin(\frac{2\pi (t-t_{0})}{P})$. This mean that $t_{0}$ can move rightside or leftside the plot. Now, let's think a little bit about this.  The starting point of a RV plot can be anywhere, it can be positive or negative. This depends on the position of the star in its orbit when the first observation is made. So, since $t_{0}$ affect this initial point, we can say that $t_{0}$ is somewhat the time at which the star is in a very special position  for observation and we can use it as a phase. This position could be the time of periaston passage or what?
